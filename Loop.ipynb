# Numpy and Matplotlib

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)
```

```python
# Define a function, e.g., y = sin(x)
y = np.sin(x)
```

```python
# Plot the graph
plt.plot(x, y, label="y = sin(x)", color="b")
```

```python
# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph of y = sin(x)")
plt.legend()
```

```python
# Show the grid
plt.grid(True)
```

```python
# Display the graph
plt.show()
```
